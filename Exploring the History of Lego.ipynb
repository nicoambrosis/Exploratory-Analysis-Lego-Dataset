## PART 1.  What percentage of all licensed sets ever released were "Star Wars" themed?

```python
len(df['theme_name'].unique())
# There are 390 theme_names
```

```python
# we are going to do a .merge with the two dfs we have
# we are going to use the merge function and we are going to use the columns df[parent_theme] and theme[name]
# parent_theme: The name of the parent theme the set belongs to. Matches the `name` column of the `parent_themes` csv file.
merged = df.merge(theme, left_on = 'parent_theme', right_on = 'name')
```

```python
merged.shape
# now we see that we have the same number of rows as the dataframe df and now we have incorporatedi the 2 extra columns 
# provided by the dataframe theme
```

```python
merged.head()
```

```python
merged[['name_x', 'name_y']].head(50)
# This block allows us to see the name_x and name_y columns in parallel and see that they do not have the same information
# What does each column mean?
```

```python
# now we see that the lego_sets database had an error in the name of the columns. The column that should be called set_name 
# was called name and that caused a problem when doing the merge. Ideally, go back and rename the column before doing 
# the merge. For the purposes of this project and to practice other things, what we are going to do is leave the merge 
# like this and change the name of the column now


merged = merged.rename(columns = {'name_x': 'set_name', 'name_y':'name'})
merged.head()


```

# PART 2.In which year was Star Wars not the most popular licensed theme? 
# (in terms of numbers of themes realesed that year)

```python
# to analyze this we would have to calculate the percentage of Star Wars for each year. We can use the groupby function
```

```python
licensed['year'].unique()
# W have data from 1999 to 2017
```

```python
licensed['Count'] = 1
# we add a column that later will allow us to easily make an account.
# This is a strategy that I learned from Keith Galli and that I find super useful, that's why I use it

summed_df = licensed.groupby(['year','parent_theme']).sum().reset_index()  
```

```python
summed_df
```

```python
max_df = summed_df.sort_values('Count', ascending=False).drop_duplicates(['year'])
max_df.sort_values('year', ascending=False)
```

```python
answer = max_df.loc[max_df['Count'] == max_df['Count'].max()]['year'].values[0]
print(f"The year Star Wars ceased to be Lego's main brand was the year {answer}")
```

## PART 3. Lets do some Descriptive statistics.
Let's make a graph showing the number of parts of the top 20 games in terms of number of parts

```python
# remove the Count column because we won't need it anymore
licensed = licensed.drop(columns = ['Count'])
licensed
```

```python
# with this block we eliminate all the rows that do not have values in the num_parts column
licensed = licensed.dropna(subset=['num_parts']).sort_values('num_parts', ascending = False).reset_index(drop = True)
licensed.head(20)
```

```python
num_parts = licensed['num_parts'].tolist()[:20]
# we are left with only the first 20 data
theme_name = licensed['theme_name'].tolist()[:20]
# we are left with only the first 20 data
```

```python
print(num_parts)
print(theme_name)
```

### Let's make the bar plot...

```python
# Make dataset
height = licensed['num_parts'].tolist()[:20] 
bars = licensed['theme_name'].tolist()[:20] 
y_pos = np.arange(len(bars))
 
# Define a specific color for Star Wars
colors = ['#ff9696' if i == 'Star Wars' else '#73c3ff' for i in bars]
    
# Create horizontal bars
plt.figure(figsize=(12, 10))

plt.barh(y_pos, height, color=colors)
 
# Create names on the y-axis
plt.yticks(y_pos, bars)
plt.xlabel("Number of pieces", fontsize=12)
plt.ylabel('Themes', fontsize=12)


# Add a line indicating Tohama ISs number
plt.axvline(x=1000, color='Black', linestyle='dashed')
plt.axvline(x=2000, color='Black', linestyle='dashed')
plt.axvline(x=3000, color='Black', linestyle='dashed')
plt.axvline(x=4000, color='Black', linestyle='dashed')
plt.axvline(x=5000, color='Black', linestyle='dashed')
 
# Save figure
plt.savefig('Lego.png', bbox_inches='tight', facecolor='white', transparent=False)    
    
# Show graphic
plt.show()
```
